```python
import numpy as np 
import pandas as pd
```

```python
df=pd.read_csv("ML_FINAL_DATA.csv")
```

```python
df
```

```python
df.drop(['Username','On a scale of 1 to 5, please rate our survey.','On a scale of 1 to 5, please rate our survey.','Unnamed: 27','Your reviews are valuable to us. Thank you once again!','Unnamed: 28','Timestamp'],axis='columns', inplace=True)
```

```python
df
```

```python
df.columns=['age','gender','famtype','dependence','wfh_rating','wfh_factor','productivity','productivityrating','stress','workquality','wfhsalary','wfhtimechange','wfhpositive','wfhnegative','wfhprefferedage','jobpreference','w-l_balance','wfhours','equiplevel','wfhessentials','wfhchallenge','wfhproductfactor','wfhlongrun']
```

```python
list(df.columns)
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['age']=le.fit_transform(df['age'])
```

```python
df['gender']=le.fit_transform(df['gender'])
```

```python
df['famtype']=le.fit_transform(df['famtype'])
```

```python
df['dependence']=le.fit_transform(df['dependence'])
```

```python
df.head()
```

```python
df.reset_index(inplace=True)
```

```python
df
```

```python
cleaned=df.set_index('index').wfh_factor.str.split(';', expand=True).stack()
```

```python
a=pd.get_dummies(cleaned, prefix='wfh_factor').groupby(level=0).sum()
```

```python
a.reset_index(inplace = True)
```

```python
a.head()
```

```python
df2=pd.merge(df, a, on="index")
```

```python
df2
```

```python
df2.drop(columns='wfh_factor',inplace=True)
```

```python
df2
```

```python
cleaned2=df.set_index('index').wfhessentials.str.split(';', expand=True).stack()
```

```python
b=pd.get_dummies(cleaned2, prefix='wfhessential').groupby(level=0).sum()
```

```python
b.reset_index(inplace = True)
```

```python
b.head()
```

```python
df3=pd.merge(df2, b, on="index")
```

```python
df3.head()
```

```python
df3['productivity']=le.fit_transform(df3['productivity'])
```

```python
df3['workquality']=le.fit_transform(df3['productivity'])
```

```python
df3.info()
```

```python
df3['wfhsalary']=le.fit_transform(df3['wfhsalary'])
```

```python
df3['wfhtimechange']=le.fit_transform(df3['wfhtimechange'])
```

```python
df3['wfhprefferedage']=le.fit_transform(df3['wfhprefferedage'])
```

```python
df3['jobpreference']=le.fit_transform(df3['jobpreference'])
```

```python
df3['w-l_balance']=le.fit_transform(df3['w-l_balance'])
```

```python
df3['wfhours']=df3['wfhours'].astype('int')
```

```python
df3.drop(columns='wfhessentials',inplace=True)
```

```python
df3.wfhchallenge=le.fit_transform(df3['wfhchallenge'])
```

```python
df3.wfhproductfactor=le.fit_transform(df3['wfhproductfactor'])
```

```python
df3.wfhlongrun=le.fit_transform(df3['wfhlongrun'])
```

```python
df3.info()
```

```python
df3.head()
```

```python
df3.drop(columns='index',inplace=True)
```

```python
df3.to_csv('processed_dataset.csv')
```
